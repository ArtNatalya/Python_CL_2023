### шаг 5

```python
def get_article_information (url):
  print (url)
  news_page = rq.get(url)
  if news_page.status_code != 200 :
    return []
  news_soup = BeautifulSoup(news_page.text, features="html.parser")
  result = []
  # заголовочки
  for h1 in news_soup.find_all('h1'):
    clazz = h1.get('class')
    if 'entry-title' in clazz :
      result.append(h1.get_text())
      break
  # авторы
  authors = []
  for a in news_soup.find_all ('a'):
    rel = a.get('rel')
    parent =  a.parent
    if rel is not None and 'author' in a.get('rel') and parent.name == 'div' and 'creators-link' in parent.get('class'):
      authors.append(a.get_text())
  all_authors = ' '.join(authors)
  result.append(all_authors)
  # дата
  for time in news_soup.find_all('time'):
    grandparent = time.parent.parent
    if 'post-meta-and-title' in grandparent.get('class'):
      result.append(time.get_text())
      break
  # текстики
  texts = []
  for p in news_soup.find_all('p'):
    parent = p.parent
    if parent.name == 'article':
     texts.append(p.get_text())
  # соединим все тексты в одну строку
  full_text = ' '.join(texts)
  result.append(full_text)
  # категории
  tags = []
  for tag in news_soup.find_all('a', rel=['category', 'tag']):
    tags.append(tag.get_text())
  all_tags = ' '.join(tags)
  result.append(all_tags)
  return result
```

```python
article_infos = []
for url in urls :
  info = get_article_information(url)
  article_infos.append(info)
print (article_infos)
```

```python
# соберите всю собранную информацию в датафрейм
article_df = pd.DataFrame(article_infos)
print(article_infos)
```
